```python
import numpy as np
from numba import cuda

# ---------------------------------
# Kernel CUDA ingênuo
# ---------------------------------

@cuda.jit
def soma_kernel(v, out):

    i = cuda.threadIdx.x

    out[i] = v[i] + v[i]


# ---------------------------------
# Programa principal
# ---------------------------------

v = np.array([1, 2, 3, 4, 5], dtype=np.int32)

out = np.zeros_like(v)

# envia para GPU
d_v = cuda.to_device(v)
d_out = cuda.to_device(out)

# executa kernel
soma_kernel[1, len(v)](d_v, d_out)

# copia resultado para CPU
resultado = d_out.copy_to_host()

print("Entrada:", v)
print("Saída:", resultado)
```
